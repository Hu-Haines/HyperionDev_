# Insurance Simple Regression

```python
# module imports
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
from sklearn.linear_model import LinearRegression
%matplotlib inline
```

```python
# importing insurance.csv and assigning it to variable insurance_data 
insurance_data = pd.read_csv('insurance.csv')
insurance_data.head()
```

### We are determining how age affects insurance costs

```python
# marking x and y vaues
# x is age and y is charges
x = insurance_data['age'].values.reshape(-1,1)
y = insurance_data['charges'].values.reshape(-1,1)
```

```python
# initial scatter plot to see whether the data is suitable
plt.scatter(x, y, color='r')
plt.xlabel('Age')
plt.ylabel('Charges')
plt.show()
```

it seems that with each age group, there are three different correlation lines,

however, we can still see a linear relationship. it could be suitable for linear regression?

```python
# fitting linear regression model on to data 
insurance_model = LinearRegression()
insurance_model.fit(x,y)
```

```python
# age 80 to add to the x-axis for prediction purpose
age_80 = [[80]]
x_with80 = np.append(x, age_80).reshape(-1,1)
```

```python
# predictions of charges 
y_pred = insurance_model.predict(x_with80)
```

```python
# plotting the model with a line of best fit
plt.scatter(x,y,color='r')
plt.plot(x_with80, y_pred, color='b')
plt.xlabel('Age')
plt.ylabel('Charges')
plt.show()
```

I think the line of best fit represents the correlation of the data.

```python
# predictions of insurance costs if ages are 20, 30 and 80
print('Prediction of insurance charge at age 20: ', insurance_model.predict([[20]]))
print('Prediction of insurance charge at age 30: ', insurance_model.predict([[30]]))
print('Prediction of insurance charge at age 80: ', insurance_model.predict(age_80))
```

we can see that as age increases the incurance charges increases aswell, and here i am assuming that the correlation doesn't change as the years goes beyond 60 (at age 80).
